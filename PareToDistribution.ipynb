### Gussian- Distribution in [Q-Q plot]

We will check whether a distribution is guassian distribution or not 
* for this we use "Q-Q plot"
* "statsmodel.api as sm"  to check whether this model is Guassian or not we use this statsmodle.api 

```python
import numpy as np
import pylab as py
import statsmodels.api as sm
```

```python
# Standard Normal Distribution --- mean =0 and standard devation =1
np.random.normal(0,1,10)   
```

```python
data_point = np.random.normal(0,1,100)    # mean, standard devaition and [100 value in b/w]
sm.qqplot(data_point,line = "45")  # 45 is the line thickness
```

* they must follow the same straight line it mean that This is "Guassian Distribution"
* Perfect Guassian Distribution must follow the straight line 

```python
import seaborn as sns
sns.displot(data_point)
```

*  Create the PDF on top of that :- "KDE" plot 

```python
sns.displot(data_point,kind = "kde")
```

### Log-Normal Distribution

```python
gfg = np.random.lognormal(0,1,50)
print(gfg)
```

```python
# this log 50 values 
```

```python
sm.qqplot(gfg, line ="45")
```

This is not Guassian distribution ,'Thus' it not follow the straight line
* Straight line point are the normal distributed point 

* log distribution ---log(x)-----> Guassian Distribution

```python
sm.qqplot(np.log(gfg),line="45")
py.show()
```

Now this Follow the straight line thus is ---> Guassian Distribution

### Pare to Distribution

pareTo it is "exponential curve" in which it different from log normal Distribution

#### Box-cox Transformation 
* of non-normal data

```python
# import modules
import numpy as np
from scipy import stats

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# generate the non-normal data(exponential)----> "Pareto"
original_data = np.random.exponential(size = 1000)

# transform the training data and save the alpha value ---> "Boxcox"
fitted_data , fitted_alpha = stats.boxcox(original_data)

# creating axes to draw plot
fig , ax = plt.subplots(1, 2)

# plotting the "Original_data" and "Fitted_data" both
sns.distplot(original_data, hist = False , kde = True,
           kde_kws = {'shade': True, 'linewidth': 2},
           label = "Non-Normal", color= "green", ax = ax[0])

sns.distplot(fitted_data, hist = False , kde = True,
           kde_kws = {'shade': True, 'linewidth': 2},
           label = "Normal", color= "green", ax = ax[1])


# adding legends to the subplots
plt.legend(loc = "upper right")


# rescaling the subplots
fig.set_figwidth(5)
fig.set_figwidth(10)

print(f"Alpha value used for Transformation  : {fitted_alpha}")
```

```python
sm.qqplot(fitted_data,line="45")
py.show()
```

#### Note:-
* This is not a "Guassian-Distribution"
* this is straight line but not falling in this line 

* Original data is "Pareto data "
* but "Fitted_data" after boxcox this is not "Gussian_Distribution"
* graph is little bit Left screww

```python
#  now apply log data 
sm.qqplot(np.log(original_data),line = "45")
py.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab

x_m = 10
alpha = 15
size = 1000000  # the size of the sample(no. of random sample)
sample = (np.random.pareto(alpha, size)+1) * x_m
stats.probplot(sample,dist="pareto", sparams=(15,10), plot = pylab)
plt.show()
```

```python
import seaborn as sns
sns.displot(sample,kind = "kde")
```

yees it is Pareto graph
