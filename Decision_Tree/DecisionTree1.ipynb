```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree 
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.metrics import accuracy_score , confusion_matrix , roc_auc_score, roc_curve
from pandas_profiling import ProfileReport
```

The dataset is "quality of wine" having mixture of different component

```python
df = pd.read_csv("winequality_red.csv")
```

```python
x = df.drop(columns="quality")
```

```python
x
```

```python
y = df["quality"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.40, random_state = 500)
```

#### Note
Accuracy is depend on the "random_state" and the "test_size" as we change the input it reflect the accuracy change

* some time duplicate row affect the accuracy
* during the checking of the score we pass the unknown data set which not pass previously 
* pass the new dataset for the "score" or "accuracy"

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(x_train,y_train)
```

```python
dt_model.predict(x_test)
```

```python
dt_model.score(x_test,y_test)  # in case as accuracy is less 57%
```

## In case change the Perameter -- 
criterion = "entropy"    the default in gini one

```python
dt_en = DecisionTreeClassifier(criterion="entropy")
```

```python
dt_en.fit(x_train,y_train)
```

```python
dt_en.score(x_test,y_test)
```
