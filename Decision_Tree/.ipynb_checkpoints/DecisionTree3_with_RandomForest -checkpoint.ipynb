## Prunning 

```python
path = dtc_model1.cost_complexity_pruning_path(x1,y1)
```

```python
path
```

```python
path = dtc_model1.cost_complexity_pruning_path(x1,y1)
ccp_alpha = path.ccp_alphas
```

here ccp_alpha is "cost_complexcity of prunning"
* cc alpha = 0 it mean infinite growing tree it will created branches
*  * now set some of "cc alpha = thershold -- then at every point it check complexity of each branch and compare it  and stop tree to grow" if value became larger

```python
ccp_alpha
```

* in case passing one by one ccp_alpha[going to each and every value]    and making the decision tree
*   default ccp_alpha = 0.0 
*  we try to bulid the decison tree with different different type        ccp_alpha value and try to test which going to fine for me [acc. to thershold]

```python
dtc_model2 = []
for ccp in ccp_alpha:
    dtc_m = DecisionTreeClassifier(ccp_alpha=ccp)
    dtc_m.fit(x1,y1)
    dtc_model2.append(dtc_m)
# we try to append our model in this list i having muliple model with different ccp_alpha value
```

```python
dtc_model2
```

```python
dtc_model2 = []
score = []
for i in ccp_alpha:
    dtc_m = DecisionTreeClassifier(ccp_alpha = i)
    dtc_m.fit(x1,y1)
    dtc_model2.append(dtc_m)
    score.append(dtc_m.score(x_test , y_test))
    
plt.plot(score, "r-+")
plt.xlabel("iteration")
plt.ylabel("score")
```

```python
dtc_model2
# how many model we build all are there with different ccp-alpha value 
# 1st one is defalut ccp_alpha=0.0
# now which i select out of this :- model which having "less overfitted" result one 
```

#  Hyper Parameter Tunning

### changing the ccp_alpha value it affect the final output 

```python
dtc_model_ccp = DecisionTreeClassifier(random_state=0, ccp_alpha =0.140)
# here we tune one perameter ccp_alpha try to perform "prunning"
# we say which branching having complexcity more than this "remove" that branch
dtc_model_ccp.fit(x1,y1)
```

## grid_pram - using random forest


```python
grid_pram = {
    "n_estimators" : [5,10 , 50 , 100 , 120 , 150],
    'criterion' :['gini' ,'entropy'],
    'max_depth' :range(10),
    'min_samples_leaf' :range(10)
    
}
```

```python
rf = RandomForestClassifier(n_estimators=5)
```

```python
grid_search_rf = GridSearchCV(param_grid = grid_pram, cv = 10, n_jobs = 6, verbose = 1, estimator = rf)
# n_jobs = 6 mean 6 core run
```

```python
# verbose = 1 using for the log functionality "use in the red color"
grid_search_rf.fit(x_train,y_train)
```

```python
grid_search_rf.best_params_
```
