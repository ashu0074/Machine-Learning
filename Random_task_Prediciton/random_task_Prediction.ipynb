### NOTE:-
* let's estimate and plot the probability density function of a random variable using Python's Matplotlib histogram function ["PDF"] - Probability Density Function 

* U(mu) -- mean and
* sigma --- Standard deviaiton 

##### Calculate the Standard Deviation and mean  

```python
import statistics
stdeviation = statistics.stdev(x)  # x=lst
meanmu = statistics.mean(x)

print("standard Deviation :",stdeviation)
print("Mean mu : ", meanmu)
```

### Gaussian - Normal Distribution

* mean+sigma --- > one standard deviation toward right (34%)
* mean-sigma ---> one standard deviation toward left(34%)
* mean(mu)-2sigma ---> 2 standard deviation toward left(42.5%)
* mean(mu)+2sigma ---> 2 standard deviation toward right(42.5%)

#### One Standard Deviation Toward Right and Left 

```python
print("One standard deviation toward Right: ",meanmu + stdeviation)
print("One standard deviation toward Left: ",meanmu - stdeviation)
```

#### Second Standard Deviation Toward Right and Left 

```python
print("Second standard deviation toward Right: ",meanmu + (2*stdeviation))
print("Second standard deviation toward Left: ", meanmu - (2 * stdeviation))
```

#### Third Standard Deviation Toward Right and Left 

```python
print("Third standard deviation toward Right: ",meanmu + (3*stdeviation))
print("Third standard deviation toward Left: ", meanmu - (3 * stdeviation))
```

#### Then I fitt normal Distribution to the data and calculate PDF

```python
from scipy import stats
Meanmu , stdeviation = stats.norm.fit(x)
points = np.linspace(stats.norm.ppf(0.01,loc = Meanmu, scale = stdeviation),
                     stats.norm.ppf(0.9999,loc= meanmu, scale= stdeviation),100)
pdf = stats.norm.pdf(points,loc = Meanmu, scale = stdeviation)
```

#### Display fitted PDF and Data Histogram

```python
plt.hist(x, 30, density=True);
plt.plot(points, pdf, color = 'r')
plt.show()
```
